```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/Pankaj-Str/Complete-Python-Mastery/refs/heads/main/47%20Seaborn/mumbai_west_housing.csv')
df.head()
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
sns.set_style('whitegrid') #clean bg
sns.set_palette('viridis') #colors
plt.rcParams['figure.figsize'] =(10,6) #size 
# ['white','dark','whitegrid','darkgrid','ticks']
```

```python
#basic distribution - price spread out
sns.histplot(data=df,x='Price_Lakhs',kde=True,color='red',bins=30)
plt.title("distributions - price spread out")
plt.xlabel("price")
plt.ylabel("count of House")
plt.show()
```

```python
# BHK most expensive ? compare price

sns.boxplot(data=df,x='BHK',y='Price_Lakhs',palette='Set2',hue='BHK')
plt.title("BHK most expensive ? compare price")
plt.xlabel("BHK")
plt.ylabel("Price (Lakhs)")
plt.show()
```

```python
#scatter plot - relationship price vs area
sns.scatterplot(data=df,
                x='Area_sqft' ,
                y='Price_Lakhs',
                hue='Locality',
                style='BHK',
                size='BHK',
                palette='tab10'
               )
plt.title("House price vs area")
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.legend(bbox_to_anchor=(1.05, 1),loc='upper left')
plt.show()
            
```
